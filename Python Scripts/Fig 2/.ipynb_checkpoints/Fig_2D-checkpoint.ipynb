```python
### import packages
import numpy as np
import dask
import dask.array as da
from glob import glob
from h5py import File
import matplotlib.pyplot as plt
import imageio
from PIL import Image
import os
import shutil
from glob import glob
import sys
import numpy as np 
%matplotlib inline

import matplotlib.pyplot as plt
%matplotlib inline

import scipy.io
from scipy.signal import find_peaks
from pandas import read_csv
```

```python
#load mean calcium and atp traces across all fish (n = 6 fish)
compiled_ca=np.load(r'G:\My Drive\adenosine_paper\adenosine_paper_data\Fig_2D\compiled_ca.npy')
compiled_atp=np.load(r'G:\My Drive\adenosine_paper\adenosine_paper_data\Fig_2D\compiled_atp.npy')
timestamp = np.load(r'G:\My Drive\adenosine_paper\adenosine_paper_data\Fig_2D\timestamp.npy')
```

```python
#calculate mean and standard error of calcium and atp
mean_ca = np.mean(compiled_ca,axis=0)
mean_atp = np.mean(compiled_atp,axis=0)
se_ca = np.std(compiled_ca,axis=0)/np.sqrt(len(compiled_ca))
se_atp = np.std(compiled_atp,axis=0)/np.sqrt(len(compiled_atp))
```

```python
#generate figure
plt.figure()
plt.plot(timestamp, mean_ca)
plt.fill_between(timestamp, mean_ca-se_ca, mean_ca+se_ca,alpha=0.5)
plt.plot(timestamp, mean_atp)
plt.fill_between(timestamp, mean_atp-se_atp, mean_atp+se_atp,alpha=0.5)
plt.axvline(x=10)
plt.savefig(r'G:\My Drive\adenosine_paper\adenosine_paper_figures\Fig2\revision\Fig_2D.svg',dpi=300)
```
